## Arrays y manipulación de datos
Los arrays son estructuras fundamentales en Julia.

```python
numeros = [1, 2, 3, 4, 5]
cuadrados = numeros .^ 2
println("Números: $numeros")
println("Cuadrados: $cuadrados")
```

## Funciones
Puedes definir tus propias funciones en Julia.

```python
function fibonacci(n)
    if n <= 1
        return n
    else
        return fibonacci(n-1) + fibonacci(n-2)
    end
end
println("Fibonacci(10) = $(fibonacci(10))")
```
