# Visualización de datos con Plots.jl
Aprende a crear gráficos en Julia usando el paquete Plots.jl.

```python
using Pkg
Pkg.add("Plots") # Ejecuta solo la primera vez
```

```python
using Plots
x = 1:10
y = x .^ 2
plot(x, y, label="y = x^2", xlabel="x", ylabel="y", title="Gráfico cuadrático")
```
